```python
import cv2
import numpy as np
from segment_anything import sam_model_registry, SamPredictor
import os
```

# 1. 모델 로드

```python
HOME = os.getcwd()
# !mkdir -p {HOME}/weights
# !wget -q https://dl.fbaipublicfiles.com/segment_anything/sam_vit_h_4b8939.pth -P {HOME}/weights
```

```python
CHECKPOINT_PATH = os.path.join(HOME, "weights", "sam_vit_h_4b8939.pth")
print(CHECKPOINT_PATH, "; exist:", os.path.isfile(CHECKPOINT_PATH))
sam = sam_model_registry["default"](checkpoint=CHECKPOINT_PATH)
predictor = SamPredictor(sam)
```

```python
def get_click(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        boxes = [np.array([x, y, x, y])]
        masks = predictor.predict(np.array(image), boxes)
        mask = masks[0].astype(np.uint8)

        # 마스크 적용하여 대상 추출
        masked_image = image * mask[:,:,None]

        # 추출 이미지 저장
        output_name = f"{image_name}_output_{x}_{y}.png"
        cv2.imwrite(output_name, masked_image)
        print(f"Extracted image saved as {output_name}")
```

```python
IMAGE_NAME = "WN001 (1).jpg"
image_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/train/nbi/" + IMAGE_NAME
image = cv2.imread(image_path)
image_name = os.path.splitext(os.path.basename(image_path))[0]
cv2.namedWindow("Image")
cv2.setMouseCallback("Image", get_click)
```
