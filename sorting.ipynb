nbi training set의 파일명의 숫자를 정하는 코드

```python
import shutil
import os
from pathlib import Path
```

```python
route = '/content/drive/MyDrive/colonoscopic/crop_datasets/nbi'

from_images = os.listdir(route+"/A")
# print(len(from_images))
for image in from_images:
  number = image[4:8]
  to_number = int(number) + 600
  dst = image.replace(str(number), str(to_number))
  os.rename(route+"/A/" + image, route+"/A/" +dst)
  # print(image, dst)
  shutil.move(route+"/A/" +dst, route)
```
