```python
ad_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/202303/adenoma"
sr_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/202303/serrated/"
hp_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/202303/hyperplasic/"
output_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/202303/datasets/"

```

```python
img_path = ad_path
#img = cv2.imread(img_path, cv2.IMREAD_COLOR
# img = Image.open(img_path)
#dst= cv2.resize(img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
folder_list = os.listdir(ad_path)
for folder in folder_list:
  NBI_path = ad_path + "/" + folder + "/NBI"
  WL_path = ad_path + "/" + folder + "/WL"
  NBI_list = os.listdir(NBI_path)
  WL_list = os.listdir(WL_path)
  num = 1
  for each_NBI in NBI_list:
    for each_WL in WL_list:
        NBI_img = cv2.imread(NBI_path + "/" + each_NBI, cv2.IMREAD_COLOR)
        WL_img = cv2.imread(WL_path + "/" + each_WL, cv2.IMREAD_COLOR)
        cv2.destroyAllWindows()

        resized_NBI = cv2.resize(NBI_img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
        resized_WL = cv2.resize(WL_img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
        cv2.destroyAllWindows()

        #가로로 붙이기
        concat_img = cv2.hconcat([resized_NBI, resized_WL])
        cv2.imwrite(output_path + "adenoma_{folder_n}_{number}.jpg".format(folder_n = folder, number=num), concat_img)
        num+=1
```
