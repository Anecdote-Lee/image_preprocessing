```python
ad_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets_202303/adenoma/"
sr_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets_202303/serrated/"
hp_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets_202303/hyperplasic/"
output_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets_202303/datasets/"
```

```python
img_path = ad_path
#img = cv2.imread(img_path, cv2.IMREAD_COLOR
# img = Image.open(img_path)
#dst= cv2.resize(img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
for i in range(40):
  NBI_path = ad_path + "NBI_" + str(i+1)
  WL_path = ad_path + "WL_" + str(i+1)

  NBI_img = cv2.imread(NBI_path, cv2.IMREAD_COLOR)
  WL_img = cv2.imread(WL_path, cv2.IMREAD_COLOR)
  cv2.destroyAllWindows()

  resized_NBI = cv2.resize(NBI_img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
  resized_WL = cv2.resize(WL_img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
  cv2.destroyAllWindows()

  #가로로 붙이기
  concat_img = cv2.hconcat([resized_NBI, resized_WL])
  cv2.imwrite(output_path + "adenoma" + str(i+1)+".jpg", concat_img)
```
