```python
from torchvision import transforms as tr
from torchvision.transforms import Compose
```

```python
def augnsave(img_route):
  for image in os.listdir(img_route):
    img = Image.open(img_route+image)
    rhf_aug = tr.RandomHorizontalFlip(p=1)(img)
    rhf_aug.save(img_route + "Flip/" + image, "png")

    rr_90_aug = tr.RandomRotation(90)(img)
    rr_90_aug.save(img_route + "Rotation/" + image, "png")

    cj_02_aug = tr.ColorJitter(brightness = 0.2, contrast = 0.2, saturation = 0.2, hue = 0.2)(img)
    cj_02_aug.save(img_route + "Colorjitter/" + image, "png")
    img.close()
  return True
```

```python
images_route = '/content/drive/MyDrive/colonoscopic/SNU_datasets/train'
nbi_hp_route = images_route + "/nbi/HP/"
wl_hp_route = images_route + "/wl/HP/"

folder_list = ['Flip', 'Rotation', 'Colorjitter']
for folder in folder_list:
    try:
        if not os.path.exists(nbi_hp_route + folder):
            os.makedirs(nbi_hp_route + folder)
    except OSError:
        print("Error: Failed to create the directory.")
    try:
        if not os.path.exists(wl_hp_route + folder):
            os.makedirs(wl_hp_route + folder)
    except OSError:
        print("Error: Failed to create the directory.")
# augnsave(nbi_hp_route)
augnsave(wl_hp_route)
```
