```python
import pandas as pd
import shutil
import os
```

```python
patients_table = pd.read_excel('/content/drive/MyDrive/colonoscopic/SNU_datasets/WN_case_list_150.xlsx', usecols=['Study No.', '시술날짜', '병리'])
```

```python
patients_table
```

```python
patients_table = patients_table.drop(['시술날짜'], axis=1)
```

```python
study_dict = patients_table.to_dict('records')
study_dict = dict(zip(patients_table['Study No.'], patients_table['병리']))
# print(study_dict)
```

```python
images_route = '/content/drive/MyDrive/colonoscopic/SNU_datasets/train'
wl_route = images_route + "/wl"
nbi_route = images_route + "/nbi"
```

```python
for nbi_image_name in os.listdir(nbi_route):
  imagename = nbi_image_name[:5]
  if study_dict[imagename] == 'AD':
    shutil.copy(nbi_route + "/" + nbi_image_name, nbi_route + "/AD/" + nbi_image_name)
  elif study_dict[imagename] == 'HP':
    shutil.copy(nbi_route + "/" + nbi_image_name, nbi_route + "/HP/" + nbi_image_name)
  else:
    pass
```

```python
for wl_image_name in os.listdir(wl_route):
  imagename = wl_image_name[:5]
  if study_dict[imagename] == 'AD':
    shutil.copy(wl_route + "/" + wl_image_name, wl_route + "/AD/" + wl_image_name)
  elif study_dict[imagename] == 'HP':
    shutil.copy(wl_route + "/" + wl_image_name, wl_route + "/HP/" + wl_image_name)
  else:
    pass
```
