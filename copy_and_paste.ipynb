```python
import shutil
import os
from pathlib import Path
```

```python
#copy, paste for train_dataset
route = '/content/drive/MyDrive/colonoscopic/SNU_datasets/202303/'

type_polyps = os.listdir(route)
wl_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/train/wl"
nbi_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/train/nbi"
adenoma_nbi = 0
adenoma_wl = 0
hyperplasic_wl = 0
hyperplasic_nbi = 0
for type in type_polyps:
  if type == 'datasets':
    continue
  else:
    folders = os.listdir(route+type)
    for folder in folders:
      folder_route = route+type+"/"+folder
      wl_route = folder_route + "/WL/"
      nbi_route = folder_route + "/NBI/"
      for wl_image in os.listdir(wl_route):
        # if os.path.isfile(wl_path + "/" + wl_image):
        #   break

        # shutil.copy(wl_route+wl_image, wl_path)
        # print(wl_route+wl_image)
        if type == "adenoma":
          adenoma_wl += 1
        elif type == "hyperplasic":
          hyperplasic_wl += 1
        else:
          pass
      for nbi_image in os.listdir(nbi_route):
        # if os.path.isfile(nbi_path + "/" + nbi_image):
        #   break
        # shutil.copy(nbi_route+nbi_image, nbi_path)
        if type == "adenoma":
          adenoma_nbi += 1
        elif type == "hyperplasic":
          hyperplasic_nbi += 1
        else:
          pass

print(adenoma_wl, adenoma_nbi)
print(hyperplasic_wl, hyperplasic_nbi)
```

```python
#copy, paste for test_dataset
test_route = '/content/drive/MyDrive/colonoscopic/open_dataset'
type_polyps = os.listdir(test_route)
test_wl_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/test/wl"
test_nbi_path = "/content/drive/MyDrive/colonoscopic/SNU_datasets/test/nbi"

adenoma_wl_cnt = adenoma_nbi_cnt = hyperplasic_nbi_cnt = hyperplasic_wl_cnt = serrated_nbi_cnt = serrated_wl_cnt = 0

for type in type_polyps:
  from_route = test_route+"/"+type + "/"
  files = os.listdir(from_route)
  if type == "adenoma":
    for file in files:
      if "WL" in file:
        adenoma_wl_cnt += 1
        # new_name = "adenoma_" + file
        # os.rename(from_route+file, from_route+new_name)
        shutil.copy(from_route+file, test_wl_path)
      elif "NBI" in file:
        adenoma_nbi_cnt += 1
        # new_name = "adenoma_" + file
        # os.rename(from_route+file, from_route+new_name)
        shutil.copy(from_route+file, test_nbi_path)
      else:
        continue
  elif type == "hyperplasic":
    for file in files:
      if "WL" in file:
        hyperplasic_wl_cnt += 1
        # new_name = "hyperplasic_" + file
        # os.rename(from_route+file, from_route+new_name)
        shutil.copy(from_route+file, test_wl_path)
      elif "NBI" in file:
        hyperplasic_nbi_cnt += 1
        # new_name = "hyperplasic_" + file
        # os.rename(from_route+file, from_route+new_name)
        shutil.copy(from_route+file, test_nbi_path)
      else:
        continue
  elif type == "serrated":
    for file in files:
      if "WL" in file:
        serrated_wl_cnt += 1
        # new_name = "serrated_" + file
        # os.rename(from_route+file, from_route+new_name)
        shutil.copy(from_route+file, test_wl_path)
      elif "NBI" in file:
        serrated_nbi_cnt += 1
        # new_name = "serrated_" + file
        # os.rename(from_route+file, from_route+new_name)
        shutil.copy(from_route+file, test_nbi_path)
      else:
        continue
  else:
    continue

print(adenoma_wl_cnt, adenoma_nbi_cnt, hyperplasic_nbi_cnt, hyperplasic_wl_cnt, serrated_nbi_cnt, serrated_wl_cnt)

```
