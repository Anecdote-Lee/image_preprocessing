```python
import pandas as pd
import shutil
import os
```

```python
students_table = pd.read_excel('/content/drive/MyDrive/colonoscopic/SNU_datasets/WN_case_list_300.xlsx', usecols=['Study No.', '시술날짜', '병리'])
```

```python
students_table
```

```python
study_table = students_table.drop(['시술날짜'], axis=1)
```

```python
study_dict = study_table.to_dict('records')
```

```python
WN_list = list()
for elem in study_dict:
  WN_list.append(elem['병리'])
```

```python
for i in range(250, len(WN_list)):
  foldername = 'WN' + "{number:0>3}".format(number=i+1)
  route = '/content/drive/MyDrive/colonoscopic/SNU_datasets/202303/'

  #WN_list에서 각 번호별로 라벨링된 병리명 폴더로 다시 정리
  if WN_list[i] == 'AD':
    shutil.move(route+foldername, route+'adenoma')
    os.mkdir(route+'adenoma/'+foldername+'/WL')
    os.mkdir(route+'adenoma/'+foldername+'/NBI')
  elif WN_list[i] == 'HP':
    shutil.move(route+foldername, route+'hyperplasic')
    os.mkdir(route+'hyperplasic/'+foldername+'/WL')
    os.mkdir(route+'hyperplasic/'+foldername+'/NBI')
  elif WN_list[i] == 'SR':
    shutil.move(route+foldername, route+'serrated')
    os.mkdir(route+'serrated/'+foldername+'/WL')
    os.mkdir(route+'serrated/'+foldername+'/NBI')
  else:
    continue
```
