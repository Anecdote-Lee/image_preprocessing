```python
import os
import pathlib
import time
import datetime
import random
import cv2
import numpy as np

from matplotlib import pyplot as plt
from IPython import display
```

```python
def generate_random_images(path, num):
  max = len(os.listdir(path+"/real_a"))
  imgs_list = range(max)
  sample_list = random.sample(imgs_list, num)
  for i in sample_list:
    wl_image = cv2.imread("{path}/real_a/sample_{num}.png".format(path = path, num=i), cv2.IMREAD_COLOR)
    nbi_image = cv2.imread("{path}/real_b/sample_{num}.png".format(path = path, num=i), cv2.IMREAD_COLOR)
    trans_image = cv2.imread("{path}/fake_b/sample_{num}.png".format(path = path, num=i), cv2.IMREAD_COLOR)
    reco_image = cv2.imread("{path}/reco_a/sample_{num}.png".format(path = path, num=i), cv2.IMREAD_COLOR)
    plt.figure(figsize=(15, 15))
    display_list = [wl_image, nbi_image, trans_image, reco_image]
    display_list = [cv2.cvtColor(out, cv2.COLOR_BGR2RGB) for out in display_list]
    title = ['Input WL Image', 'Ground Truth NBI Image', 'Translated NBI Image', 'Reconstructed WL Image']
    for i in range(4):
      plt.subplot(1, 4, i+1)
      plt.title(title[i])
      plt.imshow(display_list[i])
      plt.axis('off')
  plt.show()
```
